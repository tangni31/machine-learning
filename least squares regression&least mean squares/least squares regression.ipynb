```python
import scipy.io as scio
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
'''load training data'''
dataFile = 'training.txt'
data = scio.loadmat(dataFile)
data_training = data['training']
print (data_training)
xrange = np.arange(0,3000,1)
p1 = plt.plot(xrange, data_training, 'g')
```

```python
#using training data get w
k = 8   #order 
eta = .001 #regularization parameter 
x = np.zeros((3000-k, k))
t = np.zeros((3000-k, 1))
i = 0
while i< 3000-k:
    j = 0
    t[i] = data_training[i+k] #desire output
    while j < k:
        x[i][j] = data_training[i+j] #input
        j += 1
    i+=1
R = x.T.dot(x)/(3000-k)
p = x.T.dot(t)/(3000-k)
I = np.eye(k) 
R_new = np.array(R) + eta*I #(R+ Î»I)
R_new = R_new.astype(float)
inv_R = np.linalg.inv(R_new)
w = inv_R.dot(p)
print (w)
```
